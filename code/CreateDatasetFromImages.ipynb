creating dataset from 15 images and scores

```python
Dataset = []

path_img = 'drive/MyDrive/DL project/final/images/'

path_excel = '/content/drive/My Drive/DL project/final/excels/'

num_pictures = 15

IMDB = pd.read_csv(path_excel + 'movie_metadata.csv')
i = 0
for dir in tqdm(os.scandir(path_img)):
  #pdb.set_trace()
  score = IMDB[IMDB['movie_title'] == (dir.name+'\xa0')]['imdb_score'].values
  count = len(fnmatch.filter(os.listdir(dir), '*.*'))
  if (count > num_pictures) & (len(score) == 1) :
    files = random.sample(os.listdir(dir),num_pictures)
    tensor = []
    for file in files:
      img = cv.resize(cv.imread(path_img + dir.name +'/'+ file),(128, 128))
      tensor.append(img[:,:,0])
      tensor.append(img[:,:,1])
      tensor.append(img[:,:,2])
    Dataset.append([np.array(tensor), np.array(score)])

write_list(Dataset, 'Dataset')  
shutil.move('/content/Dataset','/content/drive/My Drive/DL project/final/Dataset')
```
