Read Datasets - Scores and URLs and creating list with this Data

```python
youtube = pd.read_csv(path + 'ml-youtube.csv')
IMDB = pd.read_csv(path + 'movie_metadata.csv')

titles = IMDB['movie_title']
years = IMDB['title_year']
scores = IMDB['imdb_score']
#print(IMDB.columns)
#url = youtube[youtube["title"] == 'Avatar (2009)']['youtubeId']
#print(url)

title_list  = []
score_list = []
url_list = []
for title, year, score in zip(titles, years, scores):
  try:
    string = title[:-1] + ' (' + str(int(year)) + ')'
    url = 'https://www.youtube.com/watch?v=' + youtube[youtube["title"] == string]['youtubeId'].values[0]
    if len(url) == 0:
      continue
    title_list.append(title)
    score_list.append(score)
    url_list.append(url)
    #pdb.set_trace()
    #print(row['youtubeId'])
  except:
    pass
```

Shuffle the Lists

```python
temp = list(zip(url_list, title_list))
np.random.shuffle(temp)
res1, res2 = zip(*temp)
# res1 and res2 come out as tuples, and so must be converted to lists.
res1, res2 = list(res1), list(res2)

```

Download a trailer, save it's interesting frames and delete the MP4 file

```python
for url, title in tqdm(zip(res1, res2)):
  try:
    real_path  = youtube_to_mp4(url, vid_path, title)
    output = save_intresting_frames(real_path, 100, 20)
    os.remove(real_path)
    if output:
      print(f'title {title} saved')
  except Exception as excep:
    print('movie ' + title + ' not saved')
    print(excep)
```
